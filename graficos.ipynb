# Graficos
Comparar os algoritmos em função no número de estados expandidos e também em função do tempo de execução.

* Para tal foram usados 5 pares de ponto inicial X ponto final diferentes para cada teste em cada um dos mapas. 
*Foram usados 5 mapas diferentes

## Busca em Largura

```python
import matplotlib.pyplot as plt

x = [16,32,64,128,256] #tamanho do mapa

y_estado_bfs = [82,429.6,2188.4,8003.4,14186.4]

plt.title('Gráfico do Estados Explorados - BFS')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Explorados')

plt.plot(x, y_estado_bfs,color="red")

plt.show()

y_tempo_bfs = [0.004,0.084764,1.69330,23.85595,88.053038]
plt.title('Gráfico de Tempo - BFS')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Tempo (s)')

plt.plot(x, y_tempo_bfs,color="red")
print()
plt.show()
```

## Busca de Custo Uniforme (UCS)

```python
x = [16,32,64,128,256] #tamanho do mapa

y_estado_ucs = [84.2,437.4,2232.4,8037,14204.4]

plt.title('Gráfico do Estados Explorados - UCS')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Explorados')

plt.plot(x, y_estado_ucs,color="blue")


plt.show()

y_tempo_ucs = [0.02137,0.092144,1.869586,24.895514,87.021612]
plt.title('Gráfico de Tempo - UCS')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Tempo (s)')

plt.plot(x, y_tempo_ucs,color="blue")
print()
plt.show()
```

Aprofundamento Iterativo (IDS)

```python
x = [16,32,64,128,256] #tamanho do mapa

y_estado_ids = [91,431.4,1626.4,7947.8,10392.8]

plt.title('Gráfico do Estados Explorados - IDS')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Explorados')

plt.plot(x, y_estado_ids,color="purple")

plt.show()

y_tempo_ids = [0.00499,0.099654,1.360676,32.382618,54.055104]
plt.title('Gráfico de Tempo - IDS')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Tempo (s)')

plt.plot(x, y_tempo_ids,color="purple")
print()
plt.show()
```

## Gulosa

```python
x = [16,32,64,128,256] #tamanho do mapa

y_estado_greedy = [13.6,33.4,168,310.4,11121]

plt.title('Gráfico do Estados Explorados - Greedy')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Explorados')

plt.plot(x, y_estado_greedy,color="hotpink")

plt.show()

y_tempo_greedy = [0.0004,0.002396,0.00302,0.136476,4.820224]
plt.title('Gráfico de Tempo - Greedy')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Tempo (s)')

plt.plot(x, y_tempo_greedy,color="hotpink")
print()
plt.show()
```

## A*

```python
x = [16,32,64,128,256] #tamanho do mapa

y_estado_astar = [36.2,148.2,1042.2,3373,9447.6]

plt.title('Gráfico do Estados Explorados - A*')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Explorados')

plt.plot(x, y_estado_astar,color="green")

plt.show()

y_tempo_astar = [0.00168,0.026978,0.60035,6.456248,47.113478]
plt.title('Gráfico de Tempo - A*')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Tempo (s)')

plt.plot(x, y_tempo_astar,color="green")
print()
plt.show()
```

## Análise Comparativa

```python
#Todos os gráficos do estados explorados juntos

plt.title('Gráfico comparativo dos Estados Explorados')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Explorados')

plt.plot(x, y_estado_astar,color="green",label="AStar")
plt.plot(x, y_estado_greedy,color="hotpink",label="Greedy")
plt.plot(x, y_estado_ids,color="purple",label="IDS")
plt.plot(x, y_estado_ucs,color="blue",label="UCS")
plt.plot(x, y_estado_bfs,color="red",label="BFS")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()

plt.show()

plt.title('Gráfico comparativo do Tempo')
plt.xlabel('Mapa (nxn)')
plt.ylabel('Tempo')

plt.plot(x, y_tempo_astar,color="green",label="AStar")
plt.plot(x, y_tempo_greedy,color="hotpink",label="Greedy")
plt.plot(x, y_tempo_ids,color="purple",label="IDS")
plt.plot(x, y_tempo_ucs,color="blue",label="UCS")
plt.plot(x, y_tempo_bfs,color="red",label="BFS")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()
print()
plt.show()
```
